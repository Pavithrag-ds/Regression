# ==========Linear Regression(Evaluation Metrics)============

# 1. Import Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2. Data Collection(Import Data)

```python
cars_data2 = pd.read_csv("Cars.csv")
cars_data2
```

# 3. Data Understanding

# 4. Data Preparation

## 4.1 Data Cleaning

## 4.2 Separate Inputs and Outputs

```python
X = cars_data2[["HP","VOL","SP","WT"]]
X
```

```python
y = cars_data2["MPG"]
y
```

# 5. Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size= 0.80, random_state= 34, shuffle= True)
X_train, X_test, y_train, y_test
```

```python
X_train
```

```python
linear_model = LinearRegression()
```

# 6. Model Training

```python
linear_model.fit(X_train,y_train)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

# 7. Model Testing

## 7.1 Training Data

```python
y_pred_train = linear_model.predict(X_train)
y_pred_train
```

## 7.2 Test Data

```python
y_pred_test = linear_model.predict(X_test)
y_pred_test
```

# 8. Model Evaluation

```python
root_mean_squared_error(y_true = y_train,y_pred = y_pred_train)  #Normalize the Units
```

```python
import statsmodels.formula.api as smf
```

```python
cars_data2.head()
```

### To see Adjusted R2 score, we are using OLS Technique from StatsModels Formula

```python
linear_stats_model = smf.ols(formula = "MPG ~ HP+VOL+SP+WT", data = cars_data2).fit()
linear_stats_model
```

```python
print("R2 Score: ",linear_stats_model.rsquared.round(4),"\nAdjusted R2 Score: ",linear_stats_model.rsquared_adj.round(4))  # \n - moveon to below line
```

```python
#R2 will be increasing when you add more features...see the below to understand this
```

```python
# only 1 input feature
linear_stats_model = smf.ols(formula = "MPG ~ HP", data = cars_data2).fit()
```

```python
print("R2 Score: ",linear_stats_model.rsquared.round(4),"\nAdjusted R2 Score: ",linear_stats_model.rsquared_adj.round(4))
```

```python
# 2 input feature
linear_stats_model = smf.ols(formula = "MPG ~ HP+VOL", data = cars_data2).fit()
```

```python
print("R2 Score: ",linear_stats_model.rsquared.round(4),"\nAdjusted R2 Score: ",linear_stats_model.rsquared_adj.round(4))
```

```python
# 3 input feature
linear_stats_model = smf.ols(formula = "MPG ~ HP+VOL+SP", data = cars_data2).fit()
```

```python
print("R2 Score: ",linear_stats_model.rsquared.round(4),"\nAdjusted R2 Score: ",linear_stats_model.rsquared_adj.round(4))
```

```python
# 4 input feature
linear_stats_model = smf.ols(formula = "MPG ~ HP+VOL+SP+WT", data = cars_data2).fit()
```

```python
print("R2 Score: ",linear_stats_model.rsquared.round(4),"\nAdjusted R2 Score: ",linear_stats_model.rsquared_adj.round(4))
```

##### Conclusion:
  - **R2 score** will ***keep increasing*** no matter how many inputs you increase and whether the added inputs contibuting more to the predictions or not.
  - But **Adjusted R2 score** will ***Not keep increasing*** if you keep increasing the inputs
    (**Adjusted R2 score** will only increase only if the added new input features is contributing more for the prediction).
  - Adding more Inputs will reduce the Error, so if  **Inputs** increases = **Residual** decreases...if **Inputs** increases = **R2** increases
